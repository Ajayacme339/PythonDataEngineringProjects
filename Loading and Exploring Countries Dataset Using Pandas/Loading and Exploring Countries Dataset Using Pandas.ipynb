This Project is for loading and Exploring Data from Countries Dataset. The data dictionary provided with data is as follows:
* Country - Name of Countries 
* Region - Region Country belongs to 
* Population - Population of Country
* Area (sq. mi.) - Area of Country
* Pop.Density (per sq. mi.) - Popultaion Density in Country
* Coastline (coast/area ratio) - Coastline Area ratio
* Net migration - Rate at which people are migrating 
* Infant mortality (per 1000 births) - Mortality rate for Infant
* GDP  - Per Capita GDP 
* Literacy(%) - Literacy rate of Country
* Phones (per 1000) - Phone per 1000 people
* Arable (%) - Arable Area
* Crops(%)- Crops Grown
* Other(%) - Other factors
* Climate - Climatic Factory for the Country 
* Birthrate - Birthrate for the Country 
* Deathrate - Deathrate for the Country 
* Agriculture - Agriculture status
* Industry - Number of Industry in the Country 
* Service - Services Offered in the Country

Dataset can be found here : https://github.com/Ajayacme339/Datacode/blob/master/countries.csv

```python
col_types = {
    col_name: countries[col_name].dtype for col_name in countries.columns
}
```

```python
countries = countries.rename(columns = {'Original Name': 'New Name',
    'Area (sq. mi.)': 'Area',
    'Pop. Density (per sq. mi.)': 'Pop Density',
    'Coastline (coast/area ratio)': 'Coastline',
    'Net migration': 'Migration',
    'Infant mortality (per 1000 births)': 'Infant Mortality',
    'GDP ($ per capita)': 'GDP Capita',
    'Literacy (%)': 'Literacy',
    'Phones (per 1000)': 'Phones',
    'Arable (%)': 'Arable',
    'Crops (%)': 'Crops',
    'Other (%)': 'Other'})
```

Column names were renamed from the countries datafarme for better readibility

```python
countries.info()
```

```python
cols_to_change = [
    'Pop Density',
    'Coastline',
    'Migration',
    'Infant Mortality',
    'Literacy',
    'Phones',
    'Arable',
    'Crops',
    'Other',
    'Climate',
    'Birthrate',
    'Deathrate',
    'Agriculture',
    'Industry',
    'Service'
]
for col in cols_to_change:
    countries[col] = countries[col].str.replace(',', '.').astype(float)
```

Data Cleaning was done for above mentioned columns in Countries Data Frame 

```python
max_pop_row = countries[countries['Population'] == countries['Population'].max()]
print(max_pop_row)
```

Country with max population was displayed and printed

```python
min_pop_row = countries[countries['Population'] == countries['Population'].min()]
print(min_pop_row)
```

Country with Min population was displayed and printed

```python
two_largest = countries.nlargest(2, columns=['Population'])
print(two_largest)
ratio = two_largest.iloc[0, 2] / two_largest.iloc[1, 2]
print(ratio)
```
