# The Purpose of this Project is to learn and apply Pandas Boolean Masks Data concepts in Data analysis and engineering for Countries Dataset(countries.csv file)

```python
#code for importing pandas and Numpy Library and reading CSV file 
import pandas as pd
import numpy as np
countries = pd.read_csv (r"C:\Users\ajaya\Downloads\countries.csv")
countries.set_index('Country',inplace = True)
                        
```

```python
#Code for renaming columns for better readibility
countries.columns = ['Region', 'Population', 'Area', 'Pop Density',
       'Coastline', 'Net migration',
       'Infant Mortality', 'GDP Capita',
       'Literacy', 'Phones', 'Arable', 'Crops',
       'Other', 'Climate', 'Birthrate', 'Deathrate', 'Agriculture',
       'Industry', 'Service']
```

```python
#Countries with birthrate less than deathrate
birth_vs_death = countries[countries['Birthrate'] < countries['Deathrate']]
print(birth_vs_death)
```

```python
#Countries with 100% Literacy
literacy_100 = countries[countries['Literacy']==100]
print(literacy_100)
```

```python
#Countries with above average gdp
above_average_gdp = countries[countries['GDP Capita']>countries['GDP Capita'].mean()]
print(above_average_gdp)
```

```python
#Countries based on agriculture more than inductry
agriculture_vs_industry =countries[countries['Agriculture']>countries['Industry']]
print(agriculture_vs_industry)
```

```python
#Countries with high density
high_density = countries[countries['Pop Density']>='10000']
print(high_density)
```

```python
#List of Countries where Population Density is high and Population Density is low
lowest_density = countries[countries['Pop Density']==countries['Pop Density'].min()].index[0]
highest_density = countries[countries['Pop Density']==countries['Pop Density'].max()].index[0]
print(lowest_density)
print(highest_density)
```

```python
#Countries with lower population desnity in Portugal
countries_lower_pt = countries[countries['Pop Density'] < countries.loc['Portugal', 'Pop Density']]
lower_density_pt_count = countries_lower_pt.shape[0]
print(lower_density_pt_count)
```

```python
#Countries where coatline is bigger than singapore
bigger_coastline_singapore_lt = countries[countries['Coastline']>countries.loc['Singapore','Coastline']]
bigger_coastline_singapore = bigger_coastline_singapore_lt.shape[0]
print(bigger_coastline_singapore)
```

```python
#Countries with more people and less land 
my_selection = countries[(countries['Population']>=100000000) & (countries['Area']<=1000000)]
print(my_selection)
```

```python

my_selection = countries[(countries['Infant Mortality']<'4') |(countries['Birthrate']>'45')]
print(my_selection)
```

```python
#Countries with Coastline areas as 0 and migration not zero
my_selection = countries[(countries['Coastline']==0) ^(countries['Migration']==0)]
print(my_selection)
```

```python
#selecting Population and Area Columns where rtegion is OCEANIA and BALTICS 
my_selection = countries.loc[(countries['Region'].isin(['OCEANIA', 'BALTICS'])) 
                             & (countries['Coastline'] <= '1'), ['Population', 'Area']]
print(my_selection)
```

```python
#Selecting Coastline,Crops,Climate Countries where region is neither WESTERN EUROPE nor EASTERN EUROPE and arable area > 50
my_selection = countries.loc[(~countries['Region'].isin(['WESTERN EUROPE','EASTERN EUROPE'])) 
                       & (countries['Arable'] > 50),['Coastline','Crops','Climate']]
                                                         
Print(my_selection)                                                       
                                                         

      
```

```python
#Selecting all the Columns from Dataframe where dataframe has missing values
cols_with_nan = countries[countries.columns[countries.isnull().any()]]
print(cols_with_nan)
```

```python
#Selecting all the numeric Columns from Data Frame
numeric_cols = countries.select_dtypes(include='number')
print(numeric_cols)
```

```python
#Seleecting all the Integer Columns from Data Frame 
integer_cols = countries.select_dtypes(include=['int64'])
print(integer_cols)
```

```python
#Selecting all the Floating Columns from Data Frame 
float_cols = countries.select_dtypes(include=['float64'])
print(float_cols)
```

```python
#Selecting all the Columns from data frame where sum is less than 1000
numeric_cols = countries.select_dtypes(include=np.number)
numeric_cols_leq_1000 = numeric_cols.loc[:, numeric_cols.sum() <= 1000]
print(numeric_cols_leq_1000)
```

```python
#Selecting all the Columns from data frame where maximum > 1000000
numeric_cols = countries.select_dtypes(include=np.number)
numeric_cols_max = numeric_cols.loc[:,numeric_cols.max()>=1000000]
print(numeric_cols_max)
```

```python
#Selecting all the Columns from data frame where values is negative
numeric_cols = countries.select_dtypes(include=np.number)
numeric_cols_negative = numeric_cols.loc[:, numeric_cols.min() < 0]
print(numeric_cols_negative)
```

```python
#Selecting all the Columns from data frame where values is between 0 and 100
numeric_cols = countries.select_dtypes(include=np.number)
cols_0_to_100 = numeric_cols.loc[:,(numeric_cols.min() >= 0) & (numeric_cols.max()<=100)]
print(cols_0_to_100)
```
